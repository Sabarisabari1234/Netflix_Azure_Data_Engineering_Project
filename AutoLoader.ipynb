#   Incremental loading using auto loader


```python
base_path = "abfss://sliver@deprojectstorage0.dfs.core.windows.net/checkpoints/v10"
schema_path = f"{base_path}/schema_metadata"
stream_checkpoint = f"{base_path}/bronze_stream"
```

```python
df = spark.readStream \
  .format("cloudFiles") \
  .option("cloudFiles.format", "csv") \
  .option("cloudFiles.schemaLocation", schema_path) \
  .load("abfss://raw@deprojectstorage0.dfs.core.windows.net/")
```

```python
df.writeStream \
  .option("checkpointLocation", stream_checkpoint) \
  .trigger(processingTime = "5 seconds") \
  .start("abfss://bronze@deprojectstorage0.dfs.core.windows.net/netflix_titles")    
```
